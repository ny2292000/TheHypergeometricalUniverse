```python
mySet=[0,1]
a=mySet[0]
b=mySet[1]
df01, autocorr01=HULib.TwoPointCorr(mySet,parameters.NumGalaxies, correctMe=parameters.correctMe)
```

# Study of density of values of z and of distances0 (distance to us).

The z profile and distance profile seems 'biased'. The question is why?  If the Universe had a bump due to acoustics 
as predicted by HU, there would naturally be a bump on the sampling. 
## ** If one were to prescribe a density of observations per value of z, then we would be BIASING the measurement**.  Sampling should pay attention only to angular coordinates and not z nor NZ (comoving number density - which I am mapping to mass or luminosity).

```python
mySet=[0,1]
a=mySet[0]
b=mySet[1]
zDensity01, dDensity01=HULib.zDistribution(mySet,parameters.NumGalaxies)
```

# zDensity and dDensity Corrected for 1/distance-squared

```python
# zDensity vs z
parameters.saveme=True
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 10))
zDensityCorrected=pd.DataFrame({'distance':zDensity01.index, 'zDensity':zDensity01.values})
zDensityCorrected= pd.DataFrame(zDensityCorrected.zDensity/zDensityCorrected.distance**2, columns=['zDensity'])
zDensityCorrected['distance']=zDensity01.index
zDensityCorrected[zDensityCorrected==np.inf]=0.0
zDensityCorrected
zDensityCorrected.plot(ax=ax, x='distance',y='zDensity', title='zDensity', ylim=[0,30])
ax.set_xlabel('z redshift')
if parameters.saveme:
    fig.savefig(parameters.imgAddress + 'zDensity01Corrected.png', dpi=300)
```

```python
parameters.saveme=True
fig=plt.figure()
dDensityCorrected=pd.DataFrame({'distance':dDensity01.index, 'dDensity':dDensity01.values})
dDensityCorrected= pd.DataFrame(dDensityCorrected.dDensity/dDensityCorrected.distance**2, columns=['dDensity'])
dDensityCorrected['distance']=dDensity01.index
dDensityCorrected[dDensityCorrected==np.inf]=0.0
fig.set_size_inches(19.5, 5.5)
dDensityCorrected.plot(x='distance',y='dDensity', title='dDensity', ylim=[0,15])
fig=plt.gcf()
plt.show()
if parameters.saveme:
    fig.savefig(parameters.imgAddress + 'dDensity01Corrected.png', dpi=200)
```

```python
# FFT of dDensity
fig, ax = plt.subplots(nrows=1, ncols=1)
y=np.fft.fft(dDensityCorrected.dDensity) 
N=y.size
dw=1/1000.0
w = np.linspace(0,N,N)*dw
plt.grid(True)
plt.plot(w,y)
# ax.set_xlim([0,1])
ax.set_ylim([-500,300])
plt.xlabel('FFT of Corrected dDensity')

fig.set_size_inches(10.5, 5.5)

if parameters.saveme:
    fig.savefig(parameters.imgAddress + 'FFT_dDensityCorrected'+str(parameters.correctMe)+'.png', dpi=200)
    
plt.show()
```

```python
# FFT of dDensity
fig, ax = plt.subplots(nrows=1, ncols=1)
y=np.fft.fft(dDensityCorrected.dDensity) 
N=y.size
dw=1/1000.0
w = np.linspace(0,N,N)*dw
plt.grid(True)
plt.plot(w,y)
ax.set_xlim([0,0.05])
ax.set_ylim([-500,300])
plt.xlabel('FFT of Corrected dDensity')

fig.set_size_inches(10.5, 5.5)

if parameters.saveme:
    fig.savefig(parameters.imgAddress + 'FFT_dDensityCorrected150Mpc'+str(parameters.correctMe)+'.png', dpi=200)
    
plt.show()
```
